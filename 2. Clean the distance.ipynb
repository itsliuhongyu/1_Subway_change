```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('dirtystations.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = df.loc[df["From"] != '起始站名']
```

```python
df['Line'] = df['Line'].map(lambda x: x.strip('路网车站站间距信息'))
df
```

Some of the lines in the original website had both to and from routes shown when my analysis only need one way. So I created a filter based on the distance and line columns. There is a risk of having two different legs with same distance and line being deleted, but I have manually checked so that it didn't happen.

```python
df2 = df.drop_duplicates(subset=["Distance", "Line"], keep='first')
df2
```

```python
df2.to_csv('cleanstations.csv')
```
