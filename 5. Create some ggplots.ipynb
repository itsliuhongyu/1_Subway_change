```python
%%R -i df

ggplot(df, aes(x=Open_Year, y=Length))+
    xlim(2000, 2022) +
    geom_bar(stat="identity", width=0.7, fill="steelblue")+
    labs(title="The largest increase in line length took place in 2010")+
    labs(subtitle="Length of Beijing Subway lines opened in each year from 2002 to 2022")+
    labs(x="Year",y="Length(km)") +
    annotate("text", x = 2010, y = 225, label = "211.1 km of lines were opened in 2010.")+
    theme_minimal()
```
